## Objective of this project is to extract data of coronavirus from worldometer by using webscraping and stored in Mysql workbench database

```python
myname = xyz.cursor()
```

```python
# Creating table to store data in database
myname.execute("CREATE TABLE CoronaPatient_Worldwide(DAY INT(2) NOT NULL AUTO_INCREMENT,REGION VARCHAR(20),TOTAL_CASES VARCHAR(20),NEW_CASES VARCHAR(20),TOTAL_DEATHS VARCHAR(20),NEW_DEATHS VARCHAR(20),TOTAL_RECOVERED VARCHAR(20),ACTIVE_CASES VARCHAR(20),SERIOUS VARCHAR(20),PRIMARY KEY (DAY))")
```

```python
url = "https://www.worldometers.info/coronavirus/"
```

```python
r = requests.get(url)
```

```python
soup = BeautifulSoup(r.content,'html.parser')
```

```python
rows = soup.find_all('tr')
```

```python
print(rows)
```

```python
s = soup.find_all('div',class_="main_table_countries_today")
```

```python
b =  soup.find_all('tr',class_="total_row_world")
```

```python
print(b[7:8])
```

```python
for row in b[7:8]:
    cols = row.find_all('td')
    cols = [x.text.strip() for x in cols]
    print(cols)
```
